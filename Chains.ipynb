```python

from langchain_groq import ChatGroq
from langchain_core.prompts import ChatPromptTemplate,HumanMessagePromptTemplate,SystemMessagePromptTemplate
```

```python
from google.colab import userdata
O = userdata.get('grok')
import os
```

```python
os.environ["GROQ_API_KEY"] = O
```

```python
from langchain_core.output_parsers import StrOutputParser
```

```python

model= ChatGroq(model = "llama-3.1-8b-instant")    #step1

prompt = ChatPromptTemplate.from_messages([HumanMessagePromptTemplate.from_template("you are supposed to generate about{topic}"),
                                           SystemMessagePromptTemplate.from_template("You are a {yr} yr experienced output generatoe")])  #step2

parser = StrOutputParser()   #3 step
```

```python

from langchain_classic.chains import LLMChain
#PT-model-parser
```

```python

chain =LLMChain(prompt = prompt,llm = model,output_parser =parser )
```

```python
a = chain.run(topic= "AI",yr = 133)
```

```python
from langchain_core.prompts import PromptTemplate
```

```python
from langchain_classic import LLMChain
```

```python
pt1 = PromptTemplate(template="you are supposed to create a story on {topic}", input_variables=["topic"])
```

```python



model1 = ChatGroq(model = "llama-3.1-8b-instant")
```

```python
parser1 = StrOutputParser()
```

```python
chain1 = LLMChain(prompt = pt1,llm=model1,output_parser=parser1, output_key = "story")
```

```python
pt2 = PromptTemplate(template="you are supposed genrate 10 quizz questions based on :{story}",input_variables="story")
```

```python
model2 = ChatGroq(model = "llama-3.1-8b-instant")
```

```python
parser2 = StrOutputParser()
```

```python
chain2 = LLMChain(prompt = pt2,llm = model2,output_parser = parser2,output_key = "mcq")
```

```python
from langchain_classic.chains import SequentialChain
```

```python
fin = SequentialChain(chains = [chain1,chain2],input_variables = ["topic"],output_variables = ["story","mcq"])
```

```python
res = fin.invoke({"topic" : "lion"})
```

```python
print(res)
```

```python
print(res['story'])
```

```python
print(res['mcq'])
```
